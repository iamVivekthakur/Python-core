```python
import numpy as np
import pandas as pd 
import seaborn as sns
```

```python
df=sns.load_dataset("iris")
df
```

```python
df.head(2)
```

```python
df.species.unique()
```

```python
df["species"]=df["species"].map({"setosa":0, "versicolor":1, "virginica":2 }) #label coding
df
```

```python
df["species"]=df["species"].replace({0:"zero", 1:"one", 2:"Two" })
df
```

```python
df1=pd.get_dummies(df["species"], drop_first=True) # one hot encoding
df1
```

```python
df=pd.concat([df,df1],axis=1)
df
```

```python
df["sq_sepal_length"]= df['sepal_length'].apply(lambda x:x**2)
df
```

```python
df["len_sp"]=df["species"].apply(lambda x : len(x))
df
```

```python
df["sq_sepal_length"]=df["sq_sepal_length"].astype("int")
df
```

```python
df.drop(columns=["species"])
df
```
